## **Basic Plotting Using Matplotlib**

Matplotlib merupakan library bawaan yang sudah ada dalam Google Colab, sehingga tidak perlu untuk menginstall Matplotlib. Sebelum memulai, mari kita import terlebih dahulu library Matplotlib. Yang harus kita import adalah matplotlib dan pyplot module. Untuk mempermudah kita dalam menulis code, matplotlib.pyplot biasanya disingkat dengan plt

```
# Import Matplotlib
import matplotlib.pyplot as plt
```

```
plt.plot([5, 10, 15, 20],[4, 9, 12, 16],color="green")
plt.show()
```

plot() function digunakan untuk membuat plot dengan element yang ada.

show() function digunakan untuk mendisplay plot pada saat kita run codenya. Function show **sangat krusial** saat menggunakan matplotlib. **Wajib** ditulis di akhir line setiap kali ingin membuat graph.

Terkadang size dari suatu grafik tidak memenuhi size yang kita inginkan. Satu cara yang bisa lakukan adalah mengadjust size tersebut mengunakan function plt.figure. Size di sini memiliki satuan inch.

```
plt.figure(figsize=(8,6))

plt.plot([5, 10, 15, 20],[4, 9, 12, 16],color="green")
plt.show()
```

Dengan fungsi plot, kita dapat menghasilkan berbagai jenis visualisasi data melalui argumen kind sebagai berikut:


*   'bar' untuk bar chart secara vertikal
*   'barh' untuk bar chart secara horizontal
*   'pie' untuk pie chart
*   'hist' untuk histogram
*   'box' untuk box plot
*   'scatter' untuk scatter plot

## **Bar Chart**

```
df_fd = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/flight_delays.csv")
df_fd.head()
```

```
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for NK Airlines Flights, by Month")
sns.barplot(x=df_fd.index, y=df_fd['NK'])
plt.ylabel("Arrival delay (in minutes)")
plt.show()
```

Insight yang bisa didapatkan dari bar chart diatas adalah bulan Mei merupakan bulan dimana total delay paling lama sepanjang satu tahun. Hal ini bisa disebabkan karena ada libur panjang di bulan Mei yaitu bulan Ramadhan.

## **Heat Map**

```
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=df_fd, annot=True, cmap="magma")
plt.xlabel("Airline")
plt.show()
```

Insight yang kita dapatkan dari heatmap diatas adalah kebanyakan pesawat lebih baik saat akhir tahun. Bisa kita lihat bahwa semakin gelap semakin bagus karena angka diatas menginformasikan tentang total delay airlines tiap bulan.

```
sns.heatmap(data=df.corr(), annot=True, cmap="Blues")
plt.show()
```

Insight yang bisa didapatkan dari heatmap diatas adalah korelasi tertinggi adalah Pclass dan Fare. Namun, korelasi nya merupakan negative yang berarti bahwa termasuk negative correlation. Semakin tinggi fare, maka semakin rendah Pclass.

## **Scatter Plot**

```
df_insurance = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/insurance.csv")
df_insurance.head()
```

```
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'])
plt.show()
```

Dari plot diatas, dapat terlihat bahwa terdapat positive correlation. Semakin tinggi bmi, maka charges akan semakin tinggi

```
sns.regplot(x=df_insurance['bmi'], y=df_insurance['charges'])
plt.show()
```

Untuk memastikan apakah data tersebut memiliki positive correlation, bisa kita check menggunakan function regplot yaitu membentuk garis regresi dari 2 feature yaitu bmi dan charges. Garis regresi diatas menunjukan bahwa terdapat positive correlation.

```
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'],hue=df_insurance['smoker'])
plt.show()
```

Dari grafik diatas, insight yang bisa kita ambil adalah non smoker cenderung untuk membayar asuransi lebih murah. Semakin tinggi bmi, smoker akan bayar asuransi semakin mahal

```
sns.lmplot(x="bmi", y="charges", hue="smoker", data=df_insurance)
plt.show()
```

lmplot() digguanakn untuk mencari garis korelasi berdasarkan hue nya. Dari sini bisa terlihat untuk smoker terdapat positive correlation yang sangat kuat. Semakin tinggi bmi, maka biaya asuransi semakin mahal. Namun, untuk non-smoker, terdapat korelasi yang sangat minimal. Terlihat bahwa garis korelasi cenderung terlihat seperti garis horizontal.

```
sns.swarmplot(x=df_insurance['smoker'], y=df_insurance['charges'])
plt.show()
```

swarmplot() digunakan untuk membuat scatterplot berdasarkan categorical variable. Dari plot diatas, bisa kita lihat bahwa secara rata-rata, non-smoker akan membayar biaya asuransi yang cenderung lebih murah daripada smoker. 

## **Histogram and Density Plot**

```
df_b = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/cancer_b.csv",index_col="Id")
df_b.head()
```

```
df_m = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/cancer_m.csv",index_col="Id")
df_m.head()
```

```
sns.distplot(a=df_b['Area (mean)'],kde=False,label='Benign Tumor')
sns.distplot(a=df_m['Area (mean)'],kde=False,label='Maligant Tumor')
plt.legend()
plt.show()
```

Insight yang dapat diambil dari graph diatas adalah benign tumor akan menyerang ke area tubuh yang jauh lebih kecil daripada maligant tumor.

```
sns.kdeplot(df['Age'],label='Age',shade=True)
plt.show()
```

kdeplot merupakan density plot. Dari graph diatas, bisa kita lihat bahwa distribusi orang yang menaiki kapal titanic terbanyak ada di usia sekitar 25 tahun.

## **Boxplot**

```
sns.boxplot(x='Pclass',y='Fare',data=df)
plt.show()
```

Dari grafik diatas, bisa kita lihat bahwa ada beberapa data yang termasuk outlier. Namun data yang terletak di Pclass 1 dengan Fare 500 sangat out of range. Untuk membuat model agar lebih akurat, biasanya data yang sangat out of range bisa kita buang.
