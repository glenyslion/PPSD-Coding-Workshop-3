## **Bar Chart**

Bar chart dapat digunakan untuk membandingkan beberapa item

```
df_fd = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/flight_delays.csv")
df_fd.head()
```

Buatlah Bar chart dari kolom NK untuk mencari tau total delay tiap bulannya

```
sns.barplot(x=df_fd['Month'], y=df_fd['NK'])
plt.show()
```

Grafik diatas terlihat kurang jelas. Judul dari grafik tidak ada, lalu size dari grafik nya perlu kita besarkan agar tidak terlalu berdempetan serta mengganti label di sumbu y.

```
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for NK Airlines Flights, by Month")
sns.barplot(x=df_fd['Month'], y=df_fd['NK'])
plt.xlabel("Month")
plt.ylabel("Arrival delay (in minutes)")
plt.show()
```

Insight yang bisa didapatkan dari bar chart diatas adalah bulan Mei merupakan bulan dimana total delay paling lama sepanjang satu tahun. Hal ini bisa disebabkan karena ada libur panjang di bulan Mei yaitu bulan Ramadhan.

## **Heat Map**

Heatmap adalah visualisasi dengan menampilkan data dengan representasi warna yang berbeda-beda. Manusia cenderung lebih mudah memahami suatu data dengan representasi visual, khususnya dengan kelompok warna yang jelas. Function yang akan kita gunakan adalah **sns.heatmap()**

Langsung aja kita coba memvisualisasi data flight delay (df_fd)

```
sns.heatmap(data=df_fd)
plt.show()
```

Disini terlihat bahwa grafik yang kita visualisasi terlihat terlalu kecil. Judul dari grafik nya juga tidak ada. Lalu susah untuk kita interpretasikan karena hanya melihat warna. 

```
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=df_fd, annot=True, cmap="magma")
plt.xlabel("Airline")
plt.show()
```

Insight yang kita dapatkan dari heatmap diatas adalah kebanyakan pesawat lebih baik saat akhir tahun. Bisa kita lihat bahwa semakin gelap semakin bagus karena angka diatas menginformasikan tentang total delay airlines tiap bulan.

Tidak hanya itu, heatmap juga bisa digunakan untuk melihat korelasi antara 2 variable.

Mari kita coba mencari korelasi dari data titanic (df)
Korelasi dapat kita cari dengan menggunakan function .corr()

```
sns.heatmap(data=df.corr())
plt.show()
```

```
sns.heatmap(data=df.corr(), annot=True, cmap="rocket_r")
plt.show()
```

Insight yang bisa didapatkan dari heatmap diatas adalah korelasi tertinggi adalah Pclass dan Fare. Namun, korelasi nya merupakan negative yang berarti bahwa termasuk negative correlation. Semakin tinggi fare, maka semakin rendah Pclass.

## **Scatter Plot**

Fungsi dari Scatter Plot adalah untuk mencari tau apakah ada korelasi antara 2/3 variable

```
df_insurance = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/insurance.csv")
df_insurance.head()
```

Disini kita ingin melihat apakah ada korelasi antara besaran bmi dengan biaya yang akan dikeluarkan untuk asuransi.

Function yang dipakai untuk membuat scatter plot adalah: **sns.scatterplot()**

```
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'])
plt.show()
```

Dari plot diatas, dapat terlihat bahwa terdapat positive correlation. Semakin tinggi bmi, maka charges akan semakin tinggi

Sekarang untuk mengetahui lebih jelas apakah data tersebut memang memiliki positive correlation, kita bisa membuat visualisasi dengan function **sns.regplot()**

```
sns.regplot(x=df_insurance['bmi'], y=df_insurance['charges'])
plt.show()
```

Selanjutnya, kita juga bisa mencari relationship dari 3 feature. Function yang kita gunakan adalah **sns.scatterplot()**. Namun, perbedaannya dengan scatterplot yang pertama, kita akan menambahkan parameter hue. Hue disini digunakan untuk grouping variable yang akan ditampilkan dalam warna yang berbeda.

Disini, kita ingin mencari tau apakah ada hubungan bmi, biaya asuransi, dan juga apakah orang tersebut merokok atau tidak.

```
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'],hue=df_insurance['smoker'])
plt.show()
```

Dari grafik diatas, insight yang bisa kita ambil adalah non smoker cenderung untuk membayar asuransi lebih murah. Semakin tinggi bmi, smoker akan bayar asuransi semakin mahal

Nah, kita perlu konfirmasi apakah intuisi yang tebak benar atau tidak. Disini kita akan check dengan membuat garis regresi pada tiap hue nya. Function yang akan kita pakai adalah **sns.lmplot()**

```
sns.lmplot(x="bmi", y="charges", hue="smoker", data=df_insurance)
plt.show()
```

Dari sini bisa terlihat untuk smoker terdapat positive correlation yang sangat kuat. Semakin tinggi bmi, maka biaya asuransi semakin mahal. Namun, untuk non-smoker, terdapat korelasi yang sangat minimal. Terlihat bahwa garis korelasi cenderung terlihat seperti garis horizontal.

Tipe terakhir untuk scatter plot adalah swarm plot. Function yang akan kita gunkanan yaitu **sns.swamplot()**. Swarm plot digunakan untuk categorical scatter plot. Categorical merupakan suatu tipe data yang bisa dikatakan dibagi menjadi ke beberapa group. Disini dapat terlihat jelas bahwa smoker merupakan categorical data.

Mari kita lihat relationship dari smoker dan charges

```
sns.swarmplot(x=df_insurance['smoker'], y=df_insurance['charges'])
plt.show()
```

Dari plot diatas, bisa kita lihat bahwa secara rata-rata, non-smoker akan membayar biaya asuransi yang cenderung lebih murah daripada smoker.

## **Histogram and Density Plot**

Mari kita load data cancer_b dan cancer_m. Disini cancer_b memiliki dari yaitu benign cancer dan cancer_m memiliki arti yaitu malignant cancer

```
df_b = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/cancer_b.csv",index_col='Id')
df_b.head()
```

```
df_m = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/cancer_m.csv",index_col="Id")
df_m.head()
```

Sekarang kita ingin mencoba untuk menggambarkan density plot sendiri. Nah untuk visualisasi density plot, function yang kita gunakan adalah **sns.kdeplot()**

kde memiliki kepanjangan yaitu kernel density estimate.

Disini kita ingin coba mencari tau distribusi dari usia dalam data Titanic (df) dengan menggunakan sns.kdeplot()

```
sns.kdeplot(df['Age'],label='Age')
plt.title("Distribution of Age")
plt.show()
```

kdeplot merupakan density plot. Dari graph diatas, bisa kita lihat bahwa distribusi orang yang menaiki kapal titanic terbanyak ada di usia sekitar 25 tahun.

Disini, kita ingin meilhat distribusi dari rata-rata area yang diserang pada saat terkena benign maupun malignant tumor. Function yang akan kita gunakan adalah **sns.distplot()**

```
sns.distplot(a=df_b['Area (mean)'])
plt.show()
```

```
sns.distplot(a=df_m['Area (mean)'])
plt.show()
```

```
sns.distplot(a=df_b['Area (mean)'])
sns.distplot(a=df_m['Area (mean)'])
plt.show()
```

Sekarang, agar tidak menggangu visualisasi dari grafik diatas, kde kita set menjadi False dan memberikan label untuk setiap plot. Serta perlu untuk kita tambahkan label dari tiap grafik.

```
sns.distplot(a=df_b['Area (mean)'],label='Benign Tumor', kde=False)
sns.distplot(a=df_m['Area (mean)'],label='Malignant Tumor',kde=False)
plt.legend()
plt.show()
```

Insight yang dapat diambil dari graph diatas adalah benign tumor akan menyerang ke area tubuh yang jauh lebih kecil daripada maligant tumor.
