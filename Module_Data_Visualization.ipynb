```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Import File DC Wikia Data
dc = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/dc-wikia-data.csv")
dc.head()
```

```
# Import File Marvel Wikia Data
marvel = pd.read_csv("https://raw.githubusercontent.com/glenyslion/PPSD-Coding-Workshop-3/main/marvel-wikia-data.csv")
marvel.head()
```

```
# Count the sex of DC
plt.figure(figsize=(10,6))
plt.title("Sex of DC")
sns.countplot(x=dc['SEX'])
plt.show()
```

Insight:

Hampir semua karakter di DC adalah male. Hanya sebagian yang female dan ada beberapa yang gendernya tidak diketahui.

```
# Count the sex of Marvel
plt.figure(figsize=(10,6))
plt.title("Sex of Marvel")
sns.countplot(x=marvel['SEX'])
plt.show()
```

Insight:

Character paling banyak adalah male. Total female hanya semabnyak 1/2 dari total male character. Ada beberapa character yang gender nya tidak diketahui.

```
# Heatmap Correlation DC
plt.figure(figsize=(10,8))
sns.heatmap(dc.corr(), annot=True)
plt.show()
```

Insight:

Di Correlation Heatmap hanya terdapat 3 variable dikarenakan correlation aku menghitung korelasi antara 2 variable yang merupakan numerical. Semua variable memiliki korelasi yang sangat rendah satu dengan yang lain.

```
# Heatmap Correlation Marvel
plt.figure(figsize=(10,8))
sns.heatmap(marvel.corr(), annot=True)
plt.show()
```

Insight:

Sama seperti heatmap dari DC, korelasi antar variable dari 3 variable diatas sangatlah rendah. Tidak ada data yang bisa kita ambil melalui grafik disini.

```
# Mencari relationship antara 3 variable (Appearances, Year, dan Align)
plt.figure(figsize=(14,7))
sns.scatterplot(x=dc['YEAR'], y=dc['APPEARANCES'], hue=dc['ALIGN'])
plt.show()
```

Insight:

Dari grafik diatas, good character akan mendapatkan jatah appearances di film jauh lebih banyak daripada character lainnya.

```
# Mencari relationship antara 3 variable (Appearances, Year, dan Align)
plt.figure(figsize=(14,7))
sns.scatterplot(x=marvel['Year'], y=marvel['APPEARANCES'], hue=marvel['ALIGN'])
plt.show()
```

Insight:

Berbeda dengan DC, good character dan neutral character banyak bermunculan di film.

```
# Melihat distribusi warna mata dengan align (character)
plt.figure(figsize=(20,5))
sns.countplot(x="ALIGN", data=dc,hue = 'EYE')
plt.title("DC")
plt.legend(loc='upper right')
plt.show()
```

Insight:

Kebanyakan superhero di DC yang merupakan good character, bad character, maupun neutral character memiliki warna mata yaitu biru.

```
# Melihat distribusi warna mata dengan align (character)
plt.figure(figsize=(20,10))
sns.countplot(x="ALIGN", data=marvel,hue = 'EYE')
plt.title("Marvel")
plt.legend(loc='upper right')
plt.show()
```

Insight:

Kebanyakan superhero di DC yang merupakan good character, bad character, maupun neutral character memiliki warna mata yaitu hazel.

```
# Distribusi pengeluaran film DC
sns.kdeplot(dc['YEAR'],label='Year')
plt.title("Distribution of Year of DC")
plt.show()
```

Insight:

Film DC kebanyakan direleasi di tahun 1980-2000an

```
# Distribusi pengeluaran film Marvel
sns.kdeplot(marvel['Year'],label='Year')
plt.title("Distribution of Year of Marvel")
plt.show()
```

Insight:

Marvel banyak diminati di tahun 1980 dengan puncak nya di tahun 1990 an dan lanjut di tahun 2010.

```
plt.figure(figsize=(20,10))
sns.boxplot(data = dc, x = 'EYE', y = 'APPEARANCES')
plt.title("Eye vs Appearances")
plt.show()
```

Insight:

Dari boxplot diatas, terlihat bahwa banyak sekali outlier. Apabila terdapat banyak outlier dapat disebabkan karena tidak ada hubungan antara warna mata dengan berapa kali appear dalam film tersebut.

```
plt.figure(figsize=(20,10))
sns.boxplot(data = dc, x = 'ALIGN', y = 'APPEARANCES')
plt.title("Align vs Appearances")
plt.show()
```

Insight:

Dari boxplot diatas, terlihat bahwa banyak sekali outlier. Apabila terdapat banyak outlier dapat disebabkan karena tidak ada hubungan antara align dengan berapa kali appear dalam film tersebut.

```
# Membuat barplot (name vs appearances)
plt.figure(figsize=(100,10))
sns.barplot(x=dc['name'],y=dc['APPEARANCES'])
plt.show()
```

Insight:

Karena data yang kita gunakan sangat banyak, figure size yang kita buat harus jauh lebih besar. Karena kategori dari nama sangatlah banyak. Namun, dengan membesarkan size grafik di atas, kita tidak bisa mengetahui nama dari superhero yang memiliki appearance paling banyak.
